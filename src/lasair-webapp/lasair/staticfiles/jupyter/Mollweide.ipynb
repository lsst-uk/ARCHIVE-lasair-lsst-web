## astropy and matplotlib for Mollweide
First import matplotlib

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

Make an astropy table

```python
from astropy.table import Table
ralist = np.array([100.0, 120.0])
delist = np.array([20.0, 30.0])
t = Table([ralist, delist], names=('RA', 'Dec'))
```

Sort out coordintes

```python
import astropy.coordinates as coord
import astropy.units as u
ra = coord.Angle(t['RA']*u.degree)
ra = ra.wrap_at(180*u.degree)
dec = coord.Angle(t['Dec']*u.degree)
```

Plot the figure

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111, projection="mollweide")
ax.grid()
ax.scatter(ra.radian, dec.radian)
```
